```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import glob

columnas = ['id', 'año', 'destino', 'estrato', 'consumo']
txt_files = glob.glob('datasets/dataset*.txt')  

for file in txt_files:
    print(f"\nAnalizando archivo: {file}")
    df = pd.read_csv(file, names=columnas, header=0, dtype=str)
    
    df['consumo'] = df['consumo'].str.replace('.', '', regex=False).str.replace(',', '.', regex=False).astype(float)
    df['estrato'] = pd.to_numeric(df['estrato'], errors='coerce')
    df['año'] = pd.to_numeric(df['año'], errors='coerce')
    
    print(df.head())
    print(df.info())
    print(df.describe())
    print(df.isnull().sum())
    print("Duplicados:", df.duplicated().sum())
    
    plt.figure(figsize=(8,3))
    df['consumo'].hist(bins=30)
    plt.title(f"Histograma consumo - {file}")
    plt.show()
    
    plt.figure(figsize=(6,2))
    df['destino'].value_counts().plot(kind='bar')
    plt.title(f"Distribución destino - {file}")
    plt.show()

    plt.figure(figsize=(8,1))
    sns.heatmap(df.isnull(), cbar=False, yticklabels=False)
    plt.title(f"Valores nulos - {file}")
    plt.show()
```

```python
df_list = []
for file in txt_files:
    df_temp = pd.read_csv(file, names=columnas, header=0, dtype=str)
    df_temp['consumo'] = df_temp['consumo'].str.replace('.', '', regex=False).str.replace(',', '.', regex=False).astype(float)
    df_temp['estrato'] = pd.to_numeric(df_temp['estrato'], errors='coerce')
    df_temp['año'] = pd.to_numeric(df_temp['año'], errors='coerce')
    df_list.append(df_temp)
    
df_all = pd.concat(df_list, ignore_index=True)
print("Tamaño total del dataset:", df_all.shape)
```

```python
print("\nResumen estadístico general:")
print(df_all.describe(include='all'))

print("\nValores nulos por columna:")
print(df_all.isnull().sum())

print("\nRegistros duplicados:", df_all.duplicated().sum())
```

```python
plt.figure(figsize=(10, 4))
sns.histplot(df_all['consumo'], bins=50, kde=True)
plt.title('Distribución global de consumo')
plt.show()

plt.figure(figsize=(8, 3))
df_all['destino'].value_counts().plot(kind='bar')
plt.title('Frecuencia global por destino')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.heatmap(df_all.corr(numeric_only=True), annot=True, cmap='Blues')
plt.title('Correlación entre variables numéricas')
plt.show()
```


Se realizó un análisis exploratorio sobre cinco archivos de micro-batch, cada uno con 500 a 10,000 registros, que parten a los consumos por diferentes destinos (Comercial, Industrial, Residencial, Oficial, Especial, Otros) y años (2023 y 2024). El objetivo fue identificar la calidad, integridad y patrones generales en los datos antes del procesamiento de la pipeline.

Estadísticas Descriptivas
Cantidad total de registros: 18,500

Cantidad de valores únicos en destino: 6 categorías

Media del consumo: Aproximadamente 912 millones

Consumo mínimo: 18,578

Consumo máximo: 1,999,997,600

Distribución de Variables
Distribución del consumo por destino:
El análisis visual con boxplots mostró que la categoría ‘Industrial’ tiende a tener los valores más altos y una mayor dispersión en el consumo, seguida por ‘Comercial’ y ‘Residencial’. Las categorías ‘Especial’ y ‘Oficial’ presentan valores medianos más bajos.
Los outliers son frecuentes en todas las categorías, evidenciando la necesidad de considerar la dispersión y valores extremos en el diseño del modelo de negocio y reglas de la pipeline.

Distribución de estrato:
La mayoría de los registros con estrato informado corresponden a valores 2 a 4, aunque su proporción es muy baja respecto al total.

Recomendaciones y Hallazgos
Mejora en la captura de datos: Se recomienda revisar la fuente de generación de los archivos para mejorar la completitud de la variable estrato, ya que su alta proporción de nulos podría limitar análisis segmentados.

Validación de outliers: Los valores atípicos y extremos en la variable consumo deben ser revisados. Dependiendo del negocio, puede ser necesario aplicar reglas para acotar el impacto de estos outliers.

Uniformidad de categorías: Se sugiere validar que las categorías de ‘destino’ sean consistentes con el catálogo de tarifas, evitando así valores inválidos en la pipeline.

Rendimiento: El procesamiento en micro-batches es adecuado y asegura un uso eficiente de la memoria, permitiendo escalar la solución a volúmenes superiores.
